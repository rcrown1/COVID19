```python
import pandas as pd
```

```python
covid = pd.read_csv('03-27-2020.txt')
```

```python
covid.Last_Update.max()
```

```python
covid_us = covid[covid['Country_Region']=='US']
```

```python
covid_us.Recovered.sum()
```

```python
covid_by_country = covid.groupby('Country_Region')[['Deaths','Recovered', 'Active','Confirmed']].sum()
```

```python
covid_by_country.sort_values('Confirmed', ascending=False)
```
